```python
fig, ax = plt.subplots(figsize=(6, 6))

A = param["R"]*param["T"]/(param["F"]*param["v"])
# courbes pour log(lambda) = 1

for i in range(9):
    n = 0.5*i - 3
    K = 10**(n)
    param["k_m"] =1/(A*(1+K))
    param["k_p"] = K/(A*(1+K))
    (param, E, C_init, M_new_constant, M_old, fun, fun_I) = initialise(param)
    print(param["k_p"])
    (I, Potential, Time) = calculate_I(param, E, C_init, M_new_constant, M_old, fun, fun_I)
    from plot_tools import extract_expe_like_CSV
    (Pot, I_extr) = extract_expe_like_CSV(param, I)
    ax.plot(Pot, np.array(I_extr)/0.0000005, label= f'log(K) = {n}', color = spectral(i))
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

ax.set_xlabel('Potential (V)')
ax.set_ylabel('Normalized Intensity (a.u.)')
ax.set_title('Voltammogram function of K for log($\lambda$) = 1')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
spectral = cm.get_cmap('Spectral', 12)
K = 10**(-0.5)
#A = param["R"]*param["T"]/(param["F"]*param["v"])
# courbes pour log(lambda) = 1

for i in range(12):
    n = 0.5*i-3
    param["k_p"] = 10**(n)
    param["k_m"] = param["k_p"]/K
    L = np.log10((param["k_p"]+param["k_m"])*param["R"]*param["T"]/(param["F"]*param["v"]))
    (param, E, C_init, M_new_constant, M_old, fun, fun_I) = initialise(param)
    (I, Potential, Time) = calculate_I(param, E, C_init, M_new_constant, M_old, fun, fun_I)
    from plot_tools import extract_expe_like_CSV
    (Pot, I_extr) = extract_expe_like_CSV(param, I)
    L = round(L,2)
    ax.plot(Pot, np.array(I_extr)/(0.0000005), label= f'k$_p$ = 10$^{{{n}}}$ cm.s$^{{-1}}$, log($\lambda$) = {L}', color = spectral(i))
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

ax.set_xlabel('Potential (V)')
ax.set_ylabel('Normalized Intensity (a.u.)')
ax.set_title('Voltammogram function of k$_p$ for log(K) = -1.0')
```
