```python
param["expe_type"] = 'LSV'
param["Ox"] = False
param["Nt"] = 5000
param["Nx"] = 200
param["L_cuve"] = 1.0
param["mechanism"] = 'EC'
param["E_red"] = -1.0
param["E_ox"] = 0.7
param["E_i"] = 0.2
param["E_0_1"] = 0
param["Delta_E"] = -0.001
param["k0"] = 0.1
(param, E, C_init, M_new_constant, M_old, fun, fun_I) = initialise(param)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

for i in range(7):
    param["kc"] = 10**(i-3)
    (param, E, C_init, M_new_constant, M_old, fun, fun_I) = initialise(param)
    (I, Potential, Time) = calculate_I(param, E, C_init, M_new_constant, M_old, fun, fun_I)
    from plot_tools import extract_expe_like_CSV
    (Pot, I_extr) = extract_expe_like_CSV(param, I)
    ax.plot(Pot, np.array(I_extr)/(0.00000005), label= 'k$_c$ = 10$^{%i}$ cm.s$^{-1}$' %(i-3), color = spectral(i))
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

ax.set_xlabel('Potentiel (V)')
ax.set_ylabel('Intensité (a.u.)')
ax.set_title('Voltammogram function of k$_c$')
```

```python
fig.savefig('EC_influence_kc', dpi = 600, bbox_inches='tight')
```
